```python
import pandas as pd
import gdown
import json
import torch
from datasets import Dataset
from transformers import AutoModelForCausalLM, AutoTokenizer, TrainingArguments, BitsAndBytesConfig
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
from trl import SFTTrainer

# ==========================================
# 0. DOWNLOAD DATA FROM GOOGLE DRIVE
# ==========================================
print("Downloading 15,000 row dataset from Google Drive...")

# Extracted the File ID from your shareable link
file_id = '1Eei7hXJ65lhCIsqEKY3AhqjIGilu1mtU'
url = f'https://drive.google.com/uc?id={file_id}'
csv_filename = 'worldsim_upgraded_dataset_15000_rows.csv'

# Download using gdown
gdown.download(url, csv_filename, quiet=False)

# ==========================================
# 1. DATA PREPROCESSING (CSV -> ChatML)
# ==========================================
print("Loading and formatting dataset...")
df = pd.read_csv(csv_filename)

# Add this line to print all column names to aid in debugging KeyError
print("DataFrame columns:", df.columns.tolist())

system_prompt = """You are the autonomous Governor and Chief Financial Officer of an Indian State.
Analyze the current macroeconomic indicators, resource supplies, and climate events.
Output ONLY valid JSON representing your optimal trade order (BID/ASK) to maximize state welfare and GDP."""

formatted_data = {"text": []}

for index, row in df.iterrows():
    input_text = (
        f"State: {row['state']} | Tick: {row['tick']}\n"
        f"--- RESOURCES ---\n"
        f"Supply -> Water: {row['water_supply']:.1f} | Food: {row['food_supply']:.1f} | Energy: {row['energy_supply']:.1f}\n"
        f"Consumption -> Water: {row['water_consumed']:.1f} | Food: {row['food_consumed']:.1f} | Energy: {row['energy_consumed']:.1f}\n"
        f"--- ECONOMY ---\n"
        f"GDP: {row['state_gdp']:.1f} | Growth: {row['gdp_growth_rate']:.2f}% | Inflation: {row['inflation_rate']:.2f}%\n"
        f"Welfare: {row['welfare_index']:.2f} | Debt-to-GDP: {row['debt_to_gdp_ratio']:.2f} | Fiscal Deficit: {row['fiscal_deficit']:.2f}\n"
        f"--- ENVIRONMENT ---\n"
        f"Active Climate Event: {row['climate_event']} (Shock Intensity: {row['shock_intensity']:.2f})\n"
        f"Formulate your market order."
    )

    output_dict = {
        "order_type": row['order_type'],
        "resource": row['resource_type'],
        "quantity": round(row['trade_quantity'], 2),
        "price": round(row['trade_price'], 2)
    }
    output_json = json.dumps(output_dict)

    chat_text = f"<|im_start|>system\n{system_prompt}<|im_end|>\n<|im_start|>user\n{input_text}<|im_end|>\n<|im_start|>assistant\n{output_json}<|im_end|>"
    formatted_data["text"].append(chat_text)

dataset = Dataset.from_dict(formatted_data)
print(f"Successfully formatted {len(dataset)} rows for training.")

# ==========================================
# 2. MODEL CONFIGURATION
# ==========================================
model_id = "Qwen/Qwen2.5-1.5B-Instruct"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
)

print("Loading Qwen Base Model...")
tokenizer = AutoTokenizer.from_pretrained(model_id)
tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    device_map="auto"
)

model = prepare_model_for_kbit_training(model)

peft_config = LoraConfig(
    r=16,
    lora_alpha=16,
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj", "gate_proj", "up_proj", "down_proj"],
    bias="none",
    task_type="CAUSAL_LM",
)
model = get_peft_model(model, peft_config)

# ==========================================
# 3. TRAINING LOOP
# ==========================================
training_args = TrainingArguments(
    output_dir="./results",
    per_device_train_batch_size=4, # Increased batch size slightly since T4 GPU can handle it with 1.5B model
    gradient_accumulation_steps=4,
    optim="paged_adamw_32bit",
    logging_steps=10,
    learning_rate=2e-4,
    fp16=True,
    num_train_epochs=3, # Will run across all 15,000 rows 3 times
    warmup_ratio=0.03,
    lr_scheduler_type="constant",
)

```

```python
# ==========================================
# 3. TRAINING LOOP
# ==========================================
# Use SFTConfig instead of standard TrainingArguments
training_args = SFTConfig(
    output_dir="./results",
    per_device_train_batch_size=4,
    gradient_accumulation_steps=4,
    optim="paged_adamw_32bit",
    logging_steps=10,
    learning_rate=2e-4,
    fp16=True,
    num_train_epochs=3,
    warmup_ratio=0.03,
    lr_scheduler_type="constant",
)

trainer = SFTTrainer(
    model=model,
    train_dataset=dataset,
    peft_config=peft_config,
    dataset_text_field="text",
    max_seq_length=2048,
    tokenizer=tokenizer,
    args=training_args,
)

print("Starting FinOps Training Phase...")
trainer.train()

# ==========================================
# 4. SAVE ADAPTERS
# ==========================================
print("Saving trained LoRA adapters...")
trainer.model.save_pretrained("finops-lora-adapter")
tokenizer.save_pretrained("finops-lora-adapter")
print("Done! Ready for GGUF conversion.")
```
